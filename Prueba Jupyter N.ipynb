```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

t=np.linspace(0,1,100)
x=(np.exp(1j*(t/2)*np.pi) - np.exp(1j*t*5*np.pi))*np.sin((3*t/2)*7*np.pi)

plt.stem(t,np.real(x))
plt.xlabel('t')
plt.ylabel('x')
plt.show()

fs=40 #fecuencia de nyquist
T=1/fs
t_n=np.arange(0,1,T)
x_n=(np.exp(1j*(t_n/2)*np.pi) - np.exp(1j*t_n*5*np.pi))*np.sin((3*t_n/2)*7*np.pi)

plt.stem(t_n,np.real(x_n))
plt.xlabel('t')
plt.ylabel('x')
plt.show()

min_len = min(len(t), len(x), len(t_n), len(x_n))
datos={'t': t[:min_len], 'x': x[:min_len], 't_n': t_n[:min_len], 'x_n': x_n[:min_len]}
df = pd.DataFrame(datos)
print(df)
```
