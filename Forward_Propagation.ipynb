```
import numpy as np
import matplotlib.pyplot as plt

x=np.array([-4,-2,0,2,4,6,8,10])
m=np.random.randint(-5,5)
b=np.random.randint(-5,5)

def fp(x,m,b):
    y=(m*x)+b
    f_y=y
    return f_y

yf=fp(x,m,b)
```

```
plt.plot(x,yf, 'red')
plt.stem(x,yf, 'black')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(y)')
plt.show()

print(f"m={int(m)}\nb={int(b)}")
print(f"yf={yf}")
```
